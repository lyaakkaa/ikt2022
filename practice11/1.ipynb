```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.model_selection as sksel
import sklearn.preprocessing as skprepro
import sklearn.linear_model as skmod


# - Open the file named “Data_to_analyse.parquet” with the read_parquet() function of the pandas package.

# - Print the dataframe.

df = pd.read_parquet(r'C:\Users\leila\OneDrive\Рабочий стол\IKT\practice11\Data_to_analyse.parquet')
print(df)


```

```python
# - You can now pass to a numpy arrays if you are more comfortable with it. To pass from a dataframe column to a numpy array (1 column), use the following instruction:

# o x1 = df[“Age”].to_numpy().reshape(-1,1)

x1 = df['Age'].to_numpy().reshape(-1,1)
x2 = df['Account_money'].to_numpy().reshape(-1,1)
y = df['Money_given'].to_numpy().reshape(-1,1)

print(x1)
print(x2)
print(y)
```

```python
# - In this dataframe, every line is a person’s information. 
# The first column is the money on the bank account of the person (in $), the second is the age of the person (in years) and the last is the amount of money given to charity by this person every year (in $/year). 
# We want to make a model to predict how much a person will give to charity per year by using its age and its amount of money on its bank account. What are the features and what is the label?
```

```python
import matplotlib.pyplot as plt


plt.scatter(x1, y)
plt.show()
plt.scatter(x2, y)
```

```python
# 2. The creation of the train and test sets

# Теперь вы разделите свои данные на две части: 
# обучающую часть, которая будет использоваться для обучения модели,
#  и тестовую часть, используемую в конце для проверки 
#  точности вашего алгоритма на невидимых данных.

# - Create two dataframes or matrix: a train 
# and a test with the function train_test_split() 
# (in the sklearn.model_selection package). 
# The size of the train dataframe (or matrix)
#  must be 80 % of the size of the complete dataframe (or matrix).

# - Print the two dataframes (or matrix)..
x = np.hstack([x1, x2])
x_train, x_test, y_train, y_test = sksel.train_test_split(x, 
                                                          y, 
                                                          train_size = 0.8,
                                                         shuffle = True)


print(x_train, x_test, y_train, y_test)


# plt.scatter(data_x_train, data_y_train)
# plt.scatter(data_x_test, data_y_test)
```

```python
# 3. The scaling of the data

# Your data have different scale. In machine learning, data with the same scale are preferred. Use the standardization to get the same scale on all your data.

# - Create one standard scaler object (StandardScaler() function of sklearn.preprocessing package)

# - Scale your features of the train set with the previously created scaler (use the fit_transform() function). The fit_transform() function accept a dataframe (or matrix). in parameter but always return a numpy matrix.

# - Do the same with the label of the train set with another standard scaler object.

# - Print your scaled features and labels.
```

```python
scaler_x = skprepro.StandardScaler()
scaler_y = skprepro.StandardScaler()
scaler_x = scaler_x.fit(x_train)
x_train_std = scaler_x.transform(x_train)
y_train_std = scaler_y.fit_transform(y_train)

print(x_train_std)
```

```python
# 4. The linear regression
# You can now use your features and label to make a linear regression model.
# - Create a linear regression model and train it with the train set (use the LinearRegression() function of the sklearn.linear_model package and the fit() function of your model)
# - Print the coefficients, the intercept and the score of your model (coef_ and intercept_ and score() fucntion).
# - Write the equation of your model.

model = skmod.LinearRegression()
model = model.fit(x_train_std, y_train_std)
print(model.coef_)
print(model.intercept_)
```

```python
# 5. The comparison with the test set.

# You can now study the accuracy of your model on unseen data.

# - Scale your features and labels of your test set by using the same scaler objects than in step 3.

# - Print the score of your model on the test set.

print(model.score(x_train_std, y_train_std))

x_test_std = scaler_x.transform(x_test)
y_test_std = scaler_y.transform(y_test)
print(model.score(x_test_std, y_test_std))
```

```python
# 6. Make predictions
# You can now use your model to make predictions.
# - Predict the amount of money given to charity by these two persons:
# o John: Money on his account: 100 000$, age: 20
# o Charles: Money on his account: 100 000 000$, age: 90
# - Remember than you have to scale your features before prediction and then do the inverse scale of the prediction after the prediction.
# - Which one will give more money to charity?

prediction_x1 = np.array([[20, 100_000], [90, 100_000_000]])
prediction_x1_std = scaler_x.transform(prediction_x1)
prediction_y1_std = model.predict(prediction_x1_std)
prediction_y1 = scaler_y.inverse_transform(prediction_y1_std)
print(prediction_y1)
```
