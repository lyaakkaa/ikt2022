```python
import pandas as pd

lists = [[] for i in range(7)]


with open(r"C:\Users\leila\OneDrive\Рабочий стол\IKT\practice2\CL.csv", "r") as file:
    csvfile = csv.reader(file)
    csvfile.__next__()
    for row in csvfile:
        for index, l in enumerate(lists):
            l.append(row[index])
#         for index in range(0, len(lists)):
#             lists[index].append(row[index])  
    for i in range(1,len(lists)):
        lists[i] = [float(j) for j in lists[i]]
    lists[0] = [datetime.datetime.strptime(i , "%Y-%m-%d") for i in lists[0]]

    dict = {}
    dict = {"Data" : lists[0], "Open" : lists[1], "High" : lists[2], "Low" : lists[3], "Close" : lists[4], "Adj close" : lists[5], "Volume" : lists[5]}


df = pd.DataFrame(dict)
# print(df)

df.columns = ["NewData", "NewOpen","NewHigh","NewLow" , "NewClose", "NewAdj close", "NewVolume" ]
df["Name"] = ["Leila" for _ in range(12448)]
df["Surname"] = ["Bekturgan" for _ in range(12448)]
date_begin = datetime.datetime(year = 2022, month = 9, day = 8)
df["Date of download"] = [date_begin for _ in range(12448)]
# print(df)


date_begin_mean = datetime.datetime(year = 2021, month = 6, day = 1)
date_end_mean = datetime.datetime(year = 2021, month = 12, day = 1)

df = df[ (df['NewData'] >= date_begin_mean) & (df['NewData'] <= date_end_mean)]


print(df)
```
